```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cd2k = pd.read_csv('200000csv_closest_distances_unparallelised_unoptimised.csv')
fd2k = pd.read_csv('200000csv_furthest_distances_unparallelised_unoptimised.csv')
```

```python
cd2kav = 0.0011209
fd2kav = 1.06761
cd2kwav = 0.00111991
fd2kwav = 0.706118
```

```python
cd2kw = pd.read_csv('200000csv_closest_distances_unparallelised_wrap.csv')
fd2kw = pd.read_csv('200000csv_furthest_distances_unparallelised_wrap.csv')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].hist(cd2k, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[0, 0].axvline(cd2kav, color='black', linestyle='--', linewidth=1.5, label=f"Avg = {cd2kav:.4f}")
axs[0, 0].set_title("Closest Distances (Serial)")
axs[0, 0].set_xlabel("Distance")
axs[0, 0].set_ylabel("Frequency")
axs[0, 0].legend()

axs[0, 1].hist(fd2k, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[0, 1].axvline(fd2kav, color='black', linestyle='--', linewidth=1.5, label=f"Avg = {fd2kav:.4f}")
axs[0, 1].set_title("Furthest Distances (Serial)")
axs[0, 1].set_xlabel("Distance")
axs[0, 1].set_ylabel("Frequency")
axs[0, 1].legend()

axs[1, 0].hist(cd2kw, bins=50, color='blue', alpha=0.7, edgecolor='black')
axs[1, 0].axvline(cd2kwav, color='black', linestyle='--', linewidth=1.5, label=f"Avg = {cd2kwav:.4f}")
axs[1, 0].set_title("Closest Distances (Wrapped)")
axs[1, 0].set_xlabel("Distance")
axs[1, 0].set_ylabel("Frequency")
axs[1, 0].legend()

axs[1, 1].hist(fd2kw, bins=50, color='red', alpha=0.7, edgecolor='black')
axs[1, 1].axvline(fd2kwav, color='black', linestyle='--', linewidth=1.5, label=f"Avg = {fd2kwav:.4f}")
axs[1, 1].set_title("Furthest Distances (Wrapped)")
axs[1, 1].set_xlabel("Distance")
axs[1, 1].set_ylabel("Frequency")
axs[1, 1].legend()

fig.suptitle("Serial vs. Wrapped Distances", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])  
plt.show()
```
